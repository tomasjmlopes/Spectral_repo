```python
folder = r'D:\\Tomas\\Rock_170x120\\PRO\\'

list_folder = natsort.natsorted([folder + f for f in os.listdir(folder) if f.endswith('.hdf5')])



new_file = h5py.File(r'D:\\Tomas\\Rock_170x120\\170x120_map_fft.h5', 'w')
map_file = r'D:\\Tomas\\Rock_170x120\\170x120_map.h5'

for i in tqdm(range(0, len(list_folder))):
    file = list_folder[i]
    hf = h5py.File(file, 'r')
    properties = [key for key in list(hf.keys()) if 'System' in key][0]
     
    if i == 0:
        scores = []
        wavelengths = []
        grp1 = new_file.create_group(properties.split(':')[0])

    scores.append(np.array(hf[properties]['fft_score']))
    wavelengths.append(np.array(hf[properties]['wavelengths']))

    # grp1_original = hf[properties]
    # print(properties)
    # grp1_original.copy(list(grp1_original.keys())[0], grp1)
    # grp1_original.copy(list(grp1_original.keys())[1], grp1)

    if list(hf.keys())[0].split(':')[1] == '0':
        sp = h5py.File(map_file, 'r')
        sample_name = list(sp.keys())[0]
        sp.copy(sample_name, new_file)
        sp.close()

    hf.close()
scores = np.array(scores).flatten()
wavelengths = np.array(wavelengths).flatten()
grp1.create_dataset('wavelengths', data = np.array(wavelengths, dtype=np.single))
grp1.create_dataset('fft_score', data = np.array(scores, dtype=np.single))

new_file.close()
```

```python
def read_data_map(file, remove_baseline = False):

    hf = h5py.File(file + '.h5', 'r')
    sample = file.split("\\")[-1]
    keys = [key for key in hf.keys()]
    sample = keys[0].split(' ')[-1]

    spectrums = np.array([np.ndarray.flatten(np.array(list(hf['Sample_ID: ' + sample]['Spot_' + str(i)]['Shot_0']['Pro']))) for i in range(0,
                                                        len(list(hf['Sample_ID: ' + sample])))])
    positions = np.array([np.ndarray.flatten(np.array(list(hf['Sample_ID: ' + sample]['Spot_' + str(i)]['position']))) for i in range(0,
                                                        len(list(hf['Sample_ID: ' + sample])))])
    wavelengths = np.array(hf['System properties']['wavelengths'])
    fft_scores = np.array(hf['System properties']['fft_score'])

    
    new_specs = []
    if remove_baseline:
        pbar = tqdm(total=len(spectrums))

        for i in range(0,len(spectrums)):
            pbar.update(1)
            spec = baseline_als(spectrums[i],100000,0.05)
            new_specs.append(spectrums[i] - spec)
    
        spectrums = np.array(new_specs)
        
    unique_x = np.unique(positions[:,0])
    unique_y = np.unique(positions[:,1])
    
    nx = len(unique_x)
    ny = len(unique_y)
    indexes = np.lexsort((positions[:,0],positions[:,1]))
    spectrums = spectrums[indexes,:]
    spectrums = np.array(spectrums.reshape((ny,nx,-1)))
    # print(spectrums.shape)
    positions = np.array(positions[indexes])
    
    hf.close()
    return spectrums, wavelengths, positions, unique_x, unique_y, fft_scores
```
