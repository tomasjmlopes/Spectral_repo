```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import k, h, c
from scipy.signal import gaussian
```

```python
from PIL import Image, ImageFilter

# Open the image file
img = Image.open(r"C:/Users/tomas/Desktop/to_upscale.png")

# Apply image enhancement filters
img1 = img.filter(ImageFilter.MedianFilter(size=3))  # Apply a median filter to denoise
img1 = img1.filter(ImageFilter.UnsharpMask(radius=3, percent=200, threshold=5))  # Apply unsharp mask to enhance~
# img1.save(r"C:/Users/tomas/Desktop/enhanced.png")
```

```python
%matplotlib widget
fig, axs = plt.subplots(1, 2)

ax = axs[0]
ax.imshow(img)
ax.grid(False)
ax.set_xlim(875, 1025)
ax.set_ylim(1825, 2025)

ax = axs[1]
ax.imshow(img1)
ax.grid(False)


fig.tight_layout()
```
